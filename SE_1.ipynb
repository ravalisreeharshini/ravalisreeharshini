```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    # Quadratic equation representing temperature change
    temperature = a * (time ** 2) + b * time + c
    return temperature

def main():
    # Hardcoded coefficients
    a_hardcoded, b_hardcoded, c_hardcoded = 0.1, -1, 30

    # Time values from 0 to 10
    time_values = np.linspace(0, 10, 50)

    # Calculate temperatures using hardcoded coefficients
    temperature_hardcoded = quadratic_model(time_values, a_hardcoded, b_hardcoded, c_hardcoded)

    # Plot the results for hardcoded coefficients
    plt.plot(time_values, temperature_hardcoded, label='Hardcoded Coefficients')

    # User input coefficients
    a_user = float(input("Enter the value of a: "))
    b_user = float(input("Enter the value of b: "))
    c_user = float(input("Enter the value of c: "))

    # Calculate temperatures using user input coefficients
    temperature_user = quadratic_model(time_values, a_user, b_user, c_user)

    # Plot the results for user input coefficients
    plt.plot(time_values, temperature_user, label='User Input Coefficients')

    # Configure plot settings
    plt.xlabel("Time")
    plt.ylabel("Temperature")
    plt.legend()
    plt.title("Weather Modeling with Quadratic Equation")
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_temp_model(time, a, b, c):

    temperature = a * time**2 + b * time + c
    return temperature

try:
  with open("3.txt","r") as file_object:
    Coefficients_sets = []
    for l in file_object:
      Coefficients = list(map(float,l.split()))
      Coefficients_sets.append(Coefficients)

except FileNotFoundError:
  print("File not found")

except ValueError:
  print("Invalid value")

# Time values for plotting
time_values = np.linspace(0, 10, 51)

# Generate plots for each set of coefficients
for Coefficients in Coefficients_sets:
  a, b, c = Coefficients
  temperature_values = quadratic_temp_model(time_values, a, b, c)
  label = f"a={a}, b={b}, c={c}"
  plt.plot(time_values, temperature_values, label=label)

# Customize plot
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.legend()
plt.title("Multiple Sets Weather Modeling")

plt.show()
```
